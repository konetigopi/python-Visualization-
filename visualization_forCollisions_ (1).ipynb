```python
## importing library
import numpy as np
import pandas as pd
```

```python
## loading the data
from google.colab import files
uploaded = files.upload()
```

```python
## data analysis
import io
dataset = pd.read_csv(io.BytesIO(uploaded['Collisions.csv']))
print(dataset.head())
print(dataset.shape)
print(dataset.info())
print(dataset.describe())
```

```python
dataset.head()
```

```python
import seaborn as sns ## importing seaborn 
import matplotlib.pyplot as plt
sns.set_theme(style="ticks", color_codes=True)
```

**Age of driver by gender**

```python
plt.figure(figsize=(12,10))   ##fig size 
sns.violinplot(x='Sex_of_Driver',y='Age_of_Driver',data=dataset, palette='Paired')  ## plotting violinplot 
plt.title("Age of driver by gender");
```

**Count that first_Point_of_Impact, Separated by Vehicle_Manoeuvre**

```python
## plotting count plot
plt.figure(figsize=(12,6))
sns.countplot(x='first_Point_of_Impact',data=dataset, palette='rainbow',hue='Vehicle_Manoeuvre')
plt.title("Count that first_Point_of_Impact, Separated by Vehicle_Manoeuvre");
```

**Count the Sex_of_Driver, Separated by Damage**

```python
plt.figure(figsize=(10,6))
sns.countplot(x='Damage',data=dataset, palette='Paired',hue='Sex_of_Driver')
plt.title("Count the Sex_of_Driver, Separated by Damage");
```

**Barplot for Road type separated by Damage**

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Damage',y="Speed_limit",data=dataset, hue='Road_Type', palette='flare')
plt.title("Barplot for Road type separated by Damage");
```

 Damages based on Urban_or_Rural_Area

```python
plt.figure(figsize=(10,6))
sns.countplot(x='Urban_or_Rural_Area',data=dataset,palette='flare',hue='Damage')
plt.title("Count the Damages based on Urban_or_Rural_Area");
```

**Count  that Light_Conditions, Separated by Damage**

```python
plt.figure(figsize=(10,6))
sns.countplot(x='Light_Conditions',data=dataset,hue='Damage',palette='flare')

plt.title("Count  that Light_Conditions, Separated by Damage");
```

**Count the Damages based on Road surface"**

```python
## countplot 
plt.figure(figsize=(10,8))
sns.countplot(x="Road_Surface_Conditions", hue="Damage",data=dataset,palette='flare')
plt.title("Count the Damages based on Road surface");
```
